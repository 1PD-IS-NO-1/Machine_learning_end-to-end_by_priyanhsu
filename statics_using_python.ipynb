```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, geom, uniform, norm

# Set a common random seed for reproducibility
np.random.seed(42)

# Number of trials
n = 50

# Sample size for plotting
x = np.arange(0, n+1)

# Binomial Distribution
p_binom = 0.5  # probability of success
binom_pmf = binom.pmf(x, n, p_binom)

# Poisson Distribution
lambda_poisson = 10  # average number of events
poisson_pmf = poisson.pmf(x, lambda_poisson)

# Geometric Distribution
p_geometric = 0.2  # probability of success on each trial
geom_pmf = geom.pmf(x, p_geometric)

# Uniform Distribution
low_uniform = 0
high_uniform = 1
uniform_pdf = uniform.pdf(x, low_uniform, high_uniform)

# Normal Distribution
mean_normal = 25
std_normal = 5
normal_pmf = norm.pdf(x, mean_normal, std_normal)

# Plotting
plt.figure(figsize=(15, 10))

plt.subplot(3, 2, 1)
plt.stem(x, binom_pmf, basefmt="b-", use_line_collection=True)
plt.title('Binomial PMF')

plt.subplot(3, 2, 2)
plt.stem(x, poisson_pmf, basefmt="b-", use_line_collection=True)
plt.title('Poisson PMF')

plt.subplot(3, 2, 3)
plt.stem(x, geom_pmf, basefmt="b-", use_line_collection=True)
plt.title('Geometric PMF')

plt.subplot(3, 2, 4)
plt.hlines(uniform_pdf, xmin=0, xmax=n, colors='b', linestyles='-', lw=2)
plt.title('Uniform PMF')

plt.subplot(3, 2, 5)
plt.stem(x, normal_pmf, basefmt="b-", use_line_collection=True)
plt.title('Normal PMF')

plt.tight_layout()
plt.show()
```

so by the help of z-score we change  use z-scores to standardize the values from other distributions to a standard normal distribution. The z-score is calculated as:


Z= (X)/


where:


X is the original random variable from the distribution,

 is the mean of the distribution,

 is the standard deviation of the distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, geom, uniform, norm

# Set a common random seed for reproducibility
np.random.seed(42)

# Number of trials
n = 50

# Sample size for plotting
x = np.arange(0, n+1)

# Points to highlight
highlight_points = [10, 25, 40]

# Binomial Distribution
p_binom = 0.5  # probability of success
binom_pmf = binom.pmf(x, n, p_binom)
binom_highlighted = binom.pmf(highlight_points, n, p_binom)

# Poisson Distribution
lambda_poisson = 10  # average number of events
poisson_pmf = poisson.pmf(x, lambda_poisson)
poisson_highlighted = poisson.pmf(highlight_points, lambda_poisson)

# Geometric Distribution
p_geometric = 0.2  # probability of success on each trial
geom_pmf = geom.pmf(x, p_geometric)
geom_highlighted = geom.pmf(highlight_points, p_geometric)

# Uniform Distribution
low_uniform = 0
high_uniform = 1
uniform_pdf = uniform.pdf(x, low_uniform, high_uniform)
uniform_highlighted = uniform.pdf(highlight_points, low_uniform, high_uniform)

# Normal Distribution
mean_normal = 25
std_normal = 5
normal_pmf = norm.pdf(x, mean_normal, std_normal)
normal_highlighted = norm.pdf(highlight_points, mean_normal, std_normal)

# Convert to z-scores
z_binom = (binom_highlighted - np.mean(binom_pmf)) / np.std(binom_pmf)
z_poisson = (poisson_highlighted - np.mean(poisson_pmf)) / np.std(poisson_pmf)
z_geom = (geom_highlighted - np.mean(geom_pmf)) / np.std(geom_pmf)
z_uniform = (uniform_highlighted - np.mean(uniform_pdf)) / np.std(uniform_pdf)
z_normal = (normal_highlighted - np.mean(normal_pmf)) / np.std(normal_pmf)

# Plotting
plt.figure(figsize=(15, 12))

# Binomial subplot
plt.subplot(3, 2, 1)
plt.stem(x, binom_pmf, basefmt="b-", use_line_collection=True, markerfmt='ro-', linefmt='b-')
plt.scatter(highlight_points, binom_highlighted, color='r')
plt.title('Binomial PMF and Histogram')
plt.subplot(3, 2, 1).hist(np.random.binomial(n, p_binom, size=1000), bins=n, density=True, alpha=0.5, color='g')

# Poisson subplot
plt.subplot(3, 2, 2)
plt.stem(x, poisson_pmf, basefmt="b-", use_line_collection=True, markerfmt='ro-', linefmt='b-')
plt.scatter(highlight_points, poisson_highlighted, color='r')
plt.title('Poisson PMF and Histogram')
plt.subplot(3, 2, 2).hist(np.random.poisson(lambda_poisson, size=1000), bins=n, density=True, alpha=0.5, color='g')

# Geometric subplot
plt.subplot(3, 2, 3)
plt.stem(x, geom_pmf, basefmt="b-", use_line_collection=True, markerfmt='ro-', linefmt='b-')
plt.scatter(highlight_points, geom_highlighted, color='r')
plt.title('Geometric PMF and Histogram')
plt.subplot(3, 2, 3).hist(np.random.geometric(p_geometric, size=1000), bins=n, density=True, alpha=0.5, color='g')

# Uniform subplot
plt.subplot(3, 2, 4)
plt.hlines(uniform.pdf(x, low_uniform, high_uniform), xmin=0, xmax=n, colors='b', linestyles='-', lw=2)
plt.scatter(highlight_points, uniform_highlighted, color='r')
plt.title('Uniform PMF and Histogram')
plt.subplot(3, 2, 4).hist(np.random.uniform(low_uniform, high_uniform, size=1000), bins=n, density=True, alpha=0.5, color='g')

# Normal subplot
plt.subplot(3, 2, 5)
plt.stem(x, normal_pmf, basefmt="b-", use_line_collection=True, markerfmt='ro-', linefmt='b-')
plt.scatter(highlight_points, normal_highlighted, color='r', label='Original Normal PDF')
```
