```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df=sn.load_dataset("tips")
```

```python
df.corr()
```

```python
sn.heatmap(df.corr())
```

```python
sn.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sn.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sn.pairplot(df)
```

```python
sn.distplot(df['tip'])
```

```python
sn.distplot(df['tip'],kde=False,bins=10)
```

```python
## Count plot

sn.countplot('sex',data=df)
```

```python
plt.hist(df.smoker)
```

```python
sn.barplot(y='total_bill',x='sex',data=df)
```

```python
sn.boxplot('smoker','total_bill', data=df)
```

```python
sn.boxplot('day','total_bill',data=df)
```

```python
sn.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sn.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```
