```python
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
## Read from specific columns
df=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1', 'COL3'])
```

```python
df.to_csv('Test.csv')
df
```
